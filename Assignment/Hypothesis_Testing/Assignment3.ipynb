```python
import pandas as pd
cutlets_data=pd.read_csv('Cutlets.csv')
cutlets_data.head()
```

```python
cutlets_data.info()
```

```python
from scipy import stats
from scipy.stats import ttest_ind
alpha=0.05
p=stats.ttest_ind(cutlets_data['Unit A'], cutlets_data['Unit B'])
if (p[1]) < alpha:
   print("There is a significant difference in the diameter of cutlets between the two units.")
else:
    print("There is no significant difference in the diameter of cutlets between the two units.")
```

```python
import pandas as pd
data=pd.read_csv('BuyerRatio.csv')
data
```

```python
from scipy import stats
from scipy.stats import chi2_contingency
alpha=0.05
data.set_index('Observed Values', inplace=True)
chi_1=stats.chi2_contingency(data)
if (chi_1[1]) < alpha:
    print("Reject the null hypothesis: There is an association between gender and region.")
else:
    print("Fail to reject the null hypothesis: There is no significant association between gender and region.")
```

```python
contingency_table = pd.crosstab(df["Phillippines"], [df["Indonesia"], df["Malta"], df["India"]])
```

```python
from scipy import stats
from scipy.stats import chi2_contingency
from scipy.stats import fisher_exact
alpha=0.05
chi=stats.chi2_contingency(contingency_table)
if (chi[1]) < 0.05:
    print("Reject the null hypothesis: There is a significant association between center and defective order forms.")
else:
    print("Fail to reject the null hypothesis: There is no significant association between center and defective order forms.")
```
