```python
import pandas as pd
replacement_map = {0: "No", 1: "Yes"}
book_df = book_df.replace(replacement_map)
df = pd.DataFrame(book_df)
df
```

```python
book_df1=pd.get_dummies(df)
book_df1
```

```python
book_df['ChildBks'].unique()
```

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
frequent_itemsets=apriori(book_df1,min_support=0.5,use_colnames=True)
frequent_itemsets
```

```python
rules=association_rules(frequent_itemsets,metric="lift",min_threshold=0.7)
rules
```

```python
numerical_features =rules.describe(include=["int64","float64"]).columns
numerical_features
```

```python
for i in numerical_features:
    skew = rules[i].skew()
    sns.distplot(rules[i], kde= False, label='Skew = %.3f' %(skew), bins=30)
    plt.legend(loc='best')
    plt.show()
```

```python
fig, axes = plt.subplots(8,1,figsize=(8,5),sharex=False,sharey=False)
sns.boxplot(x='antecedent support', data= rules,palette='crest',ax=axes[0])
sns.boxplot(x='consequent support',data= rules,palette='crest',ax=axes[1])
sns.boxplot(x='support',data= rules,palette='crest',ax=axes[2])
sns.boxplot(x='confidence',data= rules,palette='crest',ax=axes[3])
sns.boxplot(x='lift',data= rules,palette='crest',ax=axes[4])
sns.boxplot(x='leverage',data= rules,palette='crest',ax=axes[5])
sns.boxplot(x='conviction',data= rules,palette='crest',ax=axes[6])
sns.boxplot(x='zhangs_metric',data= rules,palette='crest',ax=axes[7])
plt.tight_layout(pad=0.0)
```
