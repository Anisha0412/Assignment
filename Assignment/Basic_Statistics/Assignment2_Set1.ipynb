**QUESTION-1**

```python
import numpy as np
import matplotlib.pyplot as plt
companies = ['Allied Signal','Bankers Trust','General Mills','ITT Industries','I.P Morgan&Co.','Lehman Brothers','Marriott','MCI','Merrill Lynch','Microsoft','Morgan Stanley','Sun Microsystems','Travelers','US Airways','Warner-Lambert']
Measures = [24.23,25.53,25.41,24.14,29.62,28.25,25.81,24.39,40.26,32.95,91.36,25.99,39.42,26.71,35.00]
plt.figure(figsize = (12,6))
plt.bar(companies,Measures)
plt.xticks(rotation = 90)
plt.xlabel('companies')
plt.ylabel('Measures')
plt.title('Measures for each companies')
plt.show()
Q1 = np.quantile(Measures,0.25)
Q3 = np.quantile(Measures,0.75)
IQR = Q3-Q1
Lower_Limit = Q1-1.5*IQR
Upper_Limit = Q3+1.5*IQR
outliers = [x for x in Measures if x < Lower_Limit or x > Upper_Limit]
print('Outliers:',outliers)
Mean = np.mean(Measures)
STD = np.std(Measures)
Variance = np.var(Measures)
print('Mean of measures:',Mean)
print('STD of measures:',STD)
print('Variance of measures:',Variance)
```

**QUESTION-5**

```python
probabilities = {-2000:0.1,-1000:0.1,0:0.2,1000:0.2,2000:0.3,3000:0.1}
most_likely_outcome = max(probabilities, key=probabilities.get)
print('(i) The most likely monetary outcome of the business venture is:',most_likely_outcome)
expected_value = sum(probability for outcome, probability in probabilities.items())
if expected_value > 0:
  print('(ii) The venture is likely to be successful.')
else:
  print('(ii) The venture is not likely to be successful.')
long_term_average = sum(outcome * probability for outcome, probability in probabilities.items())
print('(iii) Long-term average earnings:',long_term_average)
mean = long_term_average
variance = sum((outcome - mean) ** 2 * probability for outcome, probability in probabilities.items())
STD = variance ** 0.5
print('(iv) Measure of risk (standard deviation):',STD)
```
