**QUESTION-1**

```python
from scipy.stats import norm
Mean = 45
STD = 8
commitment_time = 60
time_until_work_begins = 10
remaining_time = commitment_time - time_until_work_begins
z_score = (remaining_time - Mean)/STD
probability = 1-norm.cdf(z_score)
print("Probability that the service manager cannot meet the commitment:", probability)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
mu = 0
sigma = 1
sample_size = 10000
x1_samples = np.random.normal(mu, sigma, sample_size)
x2_samples = np.random.normal(mu, sigma, sample_size)
two_x1 = 2 * x1_samples
x1_plus_x2 = x1_samples + x2_samples
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(two_x1, bins=30, density=True, alpha=0.7, color='red', label='2*X1')
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Distribution of 2*X1')
plt.legend()
plt.subplot(1, 2, 2)
plt.hist(x1_plus_x2, bins=30, density=True, alpha=0.7, color='blue?', label='X1+X2')
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Distribution of X1+X2')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
mean = 100
variance=20
std_dev = np.sqrt(variance)
desired_probability = 0.99
z_lower = norm.ppf((1 - desired_probability) / 2)
z_upper = norm.ppf(1 - (1 - desired_probability) / 2)
a = mean + z_lower * std_dev
b = mean + z_upper * std_dev
print("a:", a)
print("b:", b)
```

**QUESTION-5**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
conversion_rate = 45
mean_profit1 = 5
std_dev_profit1 = np.sqrt(32)
mean_profit2 = 7
std_dev_profit2 = np.sqrt(42)
alpha = 0.95
z_score = norm.ppf(1 - (1 - alpha) / 2)
profit_mean_rupees = (mean_profit1 + mean_profit2) * conversion_rate
profit_std_dev_rupees = np.sqrt(std_dev_profit1**2 + std_dev_profit2**2) * conversion_rate
lower_range = profit_mean_rupees - z_score * profit_std_dev_rupees
upper_range = profit_mean_rupees + z_score * profit_std_dev_rupees
percentile_5 = norm.ppf(0.05, loc=profit_mean_rupees, scale=profit_std_dev_rupees)
prob_loss_profit1 = norm.cdf(0, loc=mean_profit1, scale=std_dev_profit1)
prob_loss_profit2 = norm.cdf(0, loc=mean_profit2, scale=std_dev_profit2)
division_with_larger_prob_loss = "Profit1" if prob_loss_profit1 > prob_loss_profit2 else "Profit2"
print("A. Rupee range with 95% probability:", (lower_range, upper_range))
print("B. 5th percentile of profit in Rupees:", percentile_5)
print("C. Division with a larger probability of making a loss:", division_with_larger_prob_loss)
```
