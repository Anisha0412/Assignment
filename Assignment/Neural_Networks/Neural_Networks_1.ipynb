```python
from sklearn import preprocessing
label_enocoder = preprocessing.LabelEncoder()
new_df['month1'] = label_enocoder.fit_transform(new_df['month'])
new_df['day1'] = label_enocoder.fit_transform(new_df['day'])
new_df['size_category1'] = label_enocoder.fit_transform(new_df['size_category'])
```

```python
new_df.head()
```

```python
new_df.drop(columns=['month','day','size_category'],inplace=True,axis=1)
```

```python
new_df.head()
```

```python
new_df.rename(columns={'month1':'month','day1':'day','size_category1':'size_category'},inplace=True)
new_df.head()
```

```python
new_df.shape
```

```python
X = new_df.iloc[:,0:10]
Y = new_df.iloc[:,10]
```

```python
import tensorflow as tf
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
model = tf.keras.Sequential([tf.keras.layers.Dense(64, activation='relu', input_shape=(x_train.shape[1],)),tf.keras.layers.Dense(32, activation='relu'),tf.keras.layers.Dense(2, activation='softmax')])
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=10, batch_size=32, validation_split=0.2)
```

```python
test_loss, test_accuracy = model.evaluate(x_test, y_test)
print('Test Accuracy:', test_accuracy)
```
